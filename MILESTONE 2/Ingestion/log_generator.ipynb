```python
import csv
import random
from datetime import datetime, timedelta

log_levels = ["INFO", "WARNING", "ERROR"]
messages = [
    "User login successful",
    "Disk space low",
    "File not found",
    "Connection established",
    "Timeout occurred",
    "Configuration loaded",
    "Permission denied"
]

def generate_timestamp(start, count):
    return [start + timedelta(seconds=i*random.randint(1,5)) for i in range(count)]

def generate_logs(num_logs=50):
    start_time = datetime.now()
    timestamps = generate_timestamp(start_time, num_logs)
    logs = []
    for ts in timestamps:
        level = random.choices(log_levels, weights=[70, 20, 10])[0]  
        message = random.choice(messages)
        logs.append([ts.strftime("%Y-%m-%d %H:%M:%S"), level, message])
    return logs

# Save as CSV
def save_csv(logs, filename="sample_logs.csv"):
    with open(filename, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["Timestamp", "Level", "Message"])
        writer.writerows(logs)

# Save as TXT
def save_txt(logs, filename="sample_logs.txt"):
    with open(filename, "w") as f:
        for log in logs:
            f.write(f"{log[0]} [{log[1]}] {log[2]}\n")

logs = generate_logs(100)
save_csv(logs)
save_txt(logs)
print("Sample logs generated successfully!")
```

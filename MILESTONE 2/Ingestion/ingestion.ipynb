```python
import pandas as pd

def ingest_csv(file_path: str):
    """Ingest logs from a CSV file."""
    return pd.read_csv(file_path)

def ingest_json(file_path: str):
    """Ingest logs from a JSON file (lines=True for log style)."""
    return pd.read_json(file_path, lines=True)
```

```python
df_csv = ingest_csv("sample_logs.csv")
print(df_csv.head())
```

```python
import pandas as pd
import numpy as np
import dask.dataframe as dd

n_rows = 1_000_000
df_pd = pd.DataFrame({
    'id': np.arange(n_rows),
    'name': [f'Employee_{i}' for i in range(n_rows)],
    'age': np.random.randint(22, 65, size=n_rows),
    'salary': np.random.randint(40000, 150000, size=n_rows),
    'department': np.random.choice(['HR', 'Engineering', 'Marketing', 'Sales', 'Finance'], size=n_rows)
})

# Save to CSV
df_pd.to_csv('large_data.csv', index=False)
print("CSV file 'large_data.csv' created successfully.")
```
